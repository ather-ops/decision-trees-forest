# Decision Tree from Scratch - Day 2

## Conditions in Decision Trees

A condition is simply a rule used to split data at a node.

It answers a Yes/No question.

Example: is_num_legs >= 3?

Based on answer, data goes left or right.

---

## 1. Axis-Aligned vs Oblique Conditions

**Axis-Aligned Condition:** Involves only a single feature.

Example: num_legs >= 2

**Oblique Condition:** Involves multiple features.

Example: num_legs >= num_fingers

Often, decision trees are trained with only axis-aligned conditions. However, oblique splits are more powerful because they can express more complex patterns. Oblique splits sometimes produce better results at the expense of higher training and inference costs.

---

## 2. Binary vs Non-Binary Conditions

**Binary Condition:** Two possible outcomes (true or false). Decision trees containing only binary conditions are called binary decision trees.

**Non-Binary Condition:** More than two possible outcomes. These have more discriminative power than binary conditions. Decisions containing one or more non-binary conditions are called non-binary decision trees.

```python
import numpy as np
```

## Axis-Aligned vs Oblique Conditions - Hands On

```python
X = np.array([
    [2, 5],
    [4, 10],
    [6, 8]
])
```

```python
# Axis-Aligned Condition
def axis_aligned(x):
    if x[0] >= 3:
        return "right"
    else:
        return "left"
```

```python
# Oblique Condition
def oblique_condition(x):
    if x[0] + 5*x[1] >= 12:
        return "right"
    else:
        return "left"
```

```python
print("Axis-Aligned vs Oblique Conditions")
print("="*50)
print(f"{'Sample':<15} {'Axis-Aligned':<15} {'Oblique':<15}")
print("-"*50)

for x in X:
    aa = axis_aligned(x)
    ob = oblique_condition(x)
    print(f"{str(x):<15} {aa:<15} {ob:<15}")
```

```python
# Test more samples
X_test = np.array([
    [1, 1],
    [3, 2],
    [5, 1],
    [2, 3],
])

print("\nMore Test Samples:")
print("="*50)
print(f"{'Sample':<15} {'Axis-Aligned':<15} {'Oblique':<15}")
print("-"*50)

for x in X_test:
    aa = axis_aligned(x)
    ob = oblique_condition(x)
    print(f"{str(x):<15} {aa:<15} {ob:<15}")
```

## Comparison Table: Axis-Aligned vs Oblique

| Aspect | Axis-Aligned | Oblique |
|--------|--------------|---------|
| Features used | Single feature | Multiple features |
| Complexity | Low | High |
| Training cost | Low | High |
| Inference cost | Low | High |
| Expressiveness | Limited | Powerful |
| Example | num_legs >= 2 | num_legs >= num_fingers |

## Binary vs Non-Binary Conditions - Hands On

```python
# Binary Condition
def binary_condition(x):
    if x >= 3:
        return "Yes"
    else:
        return "No"
```

```python
# Non-Binary Condition
def non_binary_condition(color):
    if color == "red":
        return "Class A"
    elif color == "blue":
        return "Class B"
    else:
        return "Class C"
```

```python
print("Binary Condition Examples:")
print("="*30)
values = [1, 2, 3, 4, 5]
for v in values:
    print(f"Input: {v} -> {binary_condition(v)}")
```

```python
print("\nNon-Binary Condition Examples:")
print("="*30)
colors = ["red", "blue", "green", "yellow", "red"]
for c in colors:
    print(f"Input: {c} -> {non_binary_condition(c)}")
```

## Comparison Table: Binary vs Non-Binary

| Aspect | Binary | Non-Binary |
|--------|--------|------------|
| Number of outcomes | 2 (Yes/No) | 3 or more |
| Discriminative power | Low | High |
| Tree type | Binary Decision Tree | Non-Binary Decision Tree |
| Complexity | Simple | Complex |
| Example | x >= 3 | color == red/blue/green |
