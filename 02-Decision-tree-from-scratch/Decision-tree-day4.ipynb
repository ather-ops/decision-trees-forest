## Decision Tree Training

Decision tree is built using a recursive function that:
- Finds best condition
- Splits data
- Repeats until no split is possible

The algorithm then repeats recursively and independently on both children nodes. When no satisfying conditions are found, the node becomes a leaf. The leaf prediction is determined as the most representative label value in the examples.


---

## Step-by-Step Training Process

**Step 1:** Create a root node with a question mark

**Step 2:** Grow node #1. Find condition (example: x1 >= 1). Create two child nodes

**Step 3:** Grow node #2. If no satisfying conditions found, make the node a leaf

**Step 4:** Grow node #3. Find condition (example: x2 >= 0.5). Create two child nodes

---

## Important Concepts

**Splitter:** The routine responsible for finding the best condition. Splitters are the bottleneck when training a decision tree because they need to test many possible conditions.

**Score metrics depend on task:**
- Information gain and Gini for classification
- Mean squared error for regression

**Splitter considerations:**
- Type of features (numerical, categorical, text)
- Type of task (binary classification, multi-class, regression)
- Type of condition (threshold, in-set, oblique)
- Regularization criteria
- Memory and CPU trade-offs

## Training From Scratch - Complete Implementation

````python
import numpy as np
````

````python
# Step 1: Load Data
# Feature: x
X = np.array([0, 1, 2, 3, 4])
y = np.array(["A", "A", "B", "B", "B"])
````

````python
# Step 2: Find Best Split
def find_best_split(X, y):
    thresholds = np.unique(X)
    best_threshold = None
    best_score = -1

    for t in thresholds:
        left = y[X < t]
        right = y[X >= t]

        # simple purity check (count majority)
        score = max(np.sum(left == "A"), np.sum(left == "B")) + \
                max(np.sum(right == "A"), np.sum(right == "B"))

        if score > best_score:
            best_score = score
            best_threshold = t

    return best_threshold
````

````python
# Step 3: Recursive Tree Build
def build_tree(X, y, depth=0):
    # stop if pure
    if len(set(y)) == 1:
        return y[0]

    threshold = find_best_split(X, y)

    if threshold is None:
        return max(set(y), key=list(y).count)

    left_mask = X < threshold
    right_mask = X >= threshold

    left_tree = build_tree(X[left_mask], y[left_mask], depth+1)
    right_tree = build_tree(X[right_mask], y[right_mask], depth+1)

    return {"threshold": threshold, "left": left_tree, "right": right_tree}
````

````python
# Step 4: Train Model
tree = build_tree(X, y)
print("Trained Tree:")
print(tree)
````

````python
# Step 5: Prediction
def predict(tree, x):
    if not isinstance(tree, dict):
        return tree

    if x < tree["threshold"]:
        return predict(tree["left"], x)
    else:
        return predict(tree["right"], x)
````

````python
print("\nPredictions:")
print("="*30)
for val in X:
    print(val, "->", predict(tree, val))
````

````python
# Test new values
X_new = np.array([0.5, 1.5, 2.5, 3.5, 5])
print("\nPredictions on new values:")
print("="*30)
for val in X_new:
    print(val, "->", predict(tree, val))
````
