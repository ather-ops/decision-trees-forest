## Overfitting and Pruning

Using the algorithm described previously, we can train a decision tree that will perfectly classify training examples, assuming the examples are separable. However, if the dataset contains noise, this tree will overfit to the data and show poor test accuracy.

---

## What is Overfitting?

When a decision tree perfectly classifies all training examples but performs poorly on new data, it has overfit. This happens when the tree learns the noise in the training data rather than the underlying pattern.

**Example:** A dataset with a linear relation between feature x and label y, but with added noise. An unregularized decision tree will predict all training examples correctly but fail on new data.

---

## Regularization Methods to Limit Overfitting

Apply one or both of the following regularization criteria while training:

| Method | Description |
|--------|-------------|
| Maximum Depth | Prevent tree from growing past a certain depth (e.g., 10) |
| Minimum Examples in Leaf | Leaf with less than certain number of examples will not be split |

---

## Pruning

Pruning is a post-training regularization technique that selectively removes certain branches by converting non-leaf nodes to leaves.

**How it works:**
1. Use a validation dataset
2. Test if removing a branch improves model quality on validation data
3. If yes, prune that branch (convert to leaf)

**Note:** Using a validation dataset reduces the number of examples available for initial training.

---

## Combined Approach

Many model creators apply multiple criteria:
- Apply minimum number of examples per leaf
- Apply maximum depth to limit growth
- Prune the decision tree after training

---

## Hyperparameter Tuning with Cross-Validation

**Process:**
1. Divide training examples into p non-overlapping groups (e.g., p=10)
2. For all possible hyperparameter values (max depth, min examples)
3. Evaluate on each group using tree trained on other p-1 groups
4. Average evaluation across groups
5. Select hyperparameter with best average evaluation
6. Train final tree using all n examples with selected hyperparameters

---

## Decision Tree Interpretation

Decision trees are easily interpretable. However, changing even a few examples can completely change the structure and interpretation of the tree.

**Important:** When dataset contains many similar features, the learned tree is only one of multiple equivalent trees that fit the data.

Each leaf represents a particular corner of the dataset, making trees useful for interpreting the dataset itself.

---

## Hands-On: Overfitting vs Pruning

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Create Noisy Data
np.random.seed(42)

# Feature
X = np.linspace(0, 10, 100)

# True pattern (linear)
y_true = X

# Add noise
noise = np.random.normal(0, 2, size=100)
y = y_true + noise
```

```python
# Step 2: Split (80% train, 20% validation)
split = int(0.8 * len(X))

X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]
```

```python
# Step 3: Overfitted Model (simulate tree behavior)
# Overfit: memorize training points (like deep tree)
y_pred_train = y_train.copy()

# This is what a deep tree does: fits every noise point
```

```python
# Step 4: Pruned Model (simpler fit)
# Fit simple line (like pruned tree)
coef = np.polyfit(X_train, y_train, 1)
y_pred_pruned = np.polyval(coef, X)
```

```python
# Step 5: Plot
plt.figure(figsize=(12, 6))

plt.scatter(X_train, y_train, label="Train Data", alpha=0.5, color="blue")
plt.scatter(X_val, y_val, label="Validation Data", alpha=0.5, color="orange")

# Overfit (zig-zag style)
plt.plot(X_train, y_pred_train, label="Overfit Model (Deep Tree)", color="red", linewidth=2)

# Pruned (smooth)
plt.plot(X, y_pred_pruned, label="Pruned Model (Simple)", color="green", linewidth=2)

plt.xlabel("Feature X")
plt.ylabel("Target Y")
plt.title("Overfitting vs Pruning in Decision Trees")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Calculate errors
print("="*50)
print("Model Comparison")
print("="*50)

# Overfit model errors
train_error_overfit = np.mean((y_pred_train - y_train)**2)
print(f"Overfit Model - Train MSE: {train_error_overfit:.2f}")

# Pruned model errors
y_pred_pruned_train = np.polyval(coef, X_train)
y_pred_pruned_val = np.polyval(coef, X_val)

train_error_pruned = np.mean((y_pred_pruned_train - y_train)**2)
val_error_pruned = np.mean((y_pred_pruned_val - y_val)**2)

print(f"Pruned Model - Train MSE: {train_error_pruned:.2f}")
print(f"Pruned Model - Validation MSE: {val_error_pruned:.2f}")
print("="*50)
```

```python
# Effect of different min samples per leaf
print("\n" + "="*50)
print("Effect of Minimum Examples per Leaf")
print("="*50)

min_samples_options = [2, 5, 10]

plt.figure(figsize=(15, 4))

for idx, min_samples in enumerate(min_samples_options):
    # Simulate different levels of pruning
    if min_samples == 2:
        # Overfitted
        y_sim = y_train.copy()
        label = "min_samples=2 (Overfitted)"
        color = "red"
    elif min_samples == 5:
        # Medium
        y_sim = np.polyval(np.polyfit(X_train, y_train, 3), X_train)
        label = "min_samples=5 (Medium)"
        color = "orange"
    else:
        # Underfitted
        y_sim = np.polyval(coef, X_train)
        label = "min_samples=10 (Simpler)"
        color = "green"
    
    plt.subplot(1, 3, idx+1)
    plt.scatter(X_train, y_train, alpha=0.3, color="gray")
    plt.plot(X_train, y_sim, color=color, linewidth=2, label=label)
    plt.xlabel("Feature X")
    plt.ylabel("Target Y")
    plt.title(f"min_samples_per_leaf = {min_samples}")
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
