# Decision Tree from Scratch - Day 3

## Growing Decision Tree

Growing a decision tree is the process of building the tree by recursively splitting the dataset into smaller subsets using the best possible conditions.

A decision tree learns patterns by dividing data into smaller and more homogeneous (pure) groups.

**Pure means:** Same class in one group

---

## Solution: Heuristic Approach

Decision trees are trained using heuristics, which are approximate methods that produce near-optimal solutions efficiently.

**Example:**
- Instead of checking 100,000 options
- Pick the best option at each step

---

## Algorithm Used

Decision trees use a greedy divide-and-conquer strategy.

### Divide and Conquer
- Divide data into smaller parts
- Solve each part separately

### Greedy Approach
- At each step, the algorithm selects the best split based only on current data
- Does not think ahead
- No future planning

---

## Step-by-Step Process

1. Start with entire dataset (root node)
2. Try all possible conditions
3. Evaluate each condition using a score
4. Select the best condition
5. Split dataset into two parts
6. Repeat the process for each part
7. Stop when no good split is possible

---

## What is "Best Condition"?

Best condition is the split that gives the highest score (i.e., best separation of data).

**Score measures:** How well classes are separated

(Later: Gini / Entropy)

---

## Example (Clear Understanding)

**Dataset:**

| Species   | Bill Length |
|-----------|-------------|
| Adelie    | 18          |
| Chinstrap | 17          |
| Gentoo    | 14          |
| Gentoo    | 13          |

**Condition:** bill_length >= 16

**Result:**

| Condition | Output                    |
|-----------|---------------------------|
| Yes (>=16)| Adelie, Chinstrap         |
| No (<16)  | Gentoo, Gentoo            |

---

## Recursive Growth

After the first split, the algorithm repeats the process independently on each child node.

**Example:**

**Step 1:**
Split -> [Adelie, Chinstrap] | [Gentoo]

**Step 2:**
Now split only left part -> [Adelie] | [Chinstrap]

---

## When Does Tree Stop?

A node becomes a leaf when no further useful split is possible.

**Conditions:**
- Data is pure
- No good split found
- Stopping rule reached

---

## Leaf Node Prediction

Leaf node stores the final prediction based on the data it contains.

**Rule:** Prediction = most frequent class (majority vote)

**Example:** ["dog", "dog", "cat"] -> dog

---
