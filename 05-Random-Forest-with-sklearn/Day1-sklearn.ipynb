# Random Forest - Day 1 (Scikit-Learn)

## Definition
Random Forest is an ensemble learning algorithm that combines multiple decision trees to produce more accurate and stable predictions.

---

## Ensemble Learning
Ensemble Learning is a technique where multiple models are combined to improve overall performance.

---

## Core Idea
Instead of using one decision tree, Random Forest uses many trees and combines their predictions.

---

## Wisdom of the Crowd
The combined prediction of many models is often better than the prediction of a single model.

---

## Why Random Forest is Better than a Single Tree

| Single Decision Tree | Random Forest |
|----------------------|---------------|
| unstable | more stable |
| overfits easily | better generalization |
| sensitive to data changes | reduces overfitting |

---

## Aggregation
Aggregation means combining predictions from multiple trees.

- **Classification:** Majority Voting
- **Regression:** Average Prediction

---

## Important Condition for Good Ensemble
The individual trees should:
1. be reasonably accurate
2. be different from each other

---

## Diversity
Diversity means different trees make different mistakes and predictions.

---

## Independence
If all trees are identical, Random Forest gives no improvement.

---

## Main Goal of Random Forest
Reduce variance and improve prediction stability.

---

## One-Line Summary
**Random Forest = Multiple Decision Trees + Combined Prediction**
