# Random Forest - Day 3

## Feature Randomization

Feature Randomization means each decision tree in the random forest only gets to see a random subset of features at each split.

---

## Why Feature Randomization?

Without feature randomization, most trees would look similar and make similar mistakes.

---

## How Feature Randomization Works

At each node, instead of checking all features, the tree only checks a random subset.

---

## Typical Number of Features

| Task | Features to consider |
|------|---------------------|
| Classification | sqrt(total_features) |
| Regression | total_features / 3 |

---

## Example

Dataset has 10 features:
- Classification: try sqrt(10) ≈ 3 random features at each split
- Regression: try 10/3 ≈ 3 random features at each split

---

## Why Randomization Helps

| Without Feature Randomization | With Feature Randomization |
|------------------------------|---------------------------|
| All trees use same strong feature | Different trees use different features |
| Trees become similar | Trees become diverse |
| High correlation between trees | Low correlation between trees |

---

## Two Sources of Randomness in Random Forest

1. **Bagging:** Each tree gets different training data (with replacement)
2. **Feature Randomization:** Each tree gets different features at each split

---

## Benefit of Feature Randomization

- If one feature is very strong, all trees would use it
- Then all trees would make similar predictions
- Randomization forces trees to look at other features too
- This creates more diverse trees

---

## One-Line Summary
**Feature Randomization forces each tree to consider different random features at each split.**

```python
import numpy as np
import math
```

## Feature Randomization Simulation

```python
# Total features in dataset
total_features = 10

# For classification: sqrt(total_features)
features_for_split = int(math.sqrt(total_features))

print(f"Total features: {total_features}")
print(f"Features to consider at each split: {features_for_split}")
```

```python
# Simulate random feature selection at a split
all_features = list(range(total_features))
random_features = np.random.choice(all_features, size=features_for_split, replace=False)

print(f"\nAll features: {all_features}")
print(f"Randomly selected features for this split: {sorted(random_features)}")
```

```python
# Simulate different trees seeing different features
print("\n" + "="*50)
print("Different Trees - Different Random Features")
print("="*50)

for tree in range(5):
    random_features = np.random.choice(all_features, size=features_for_split, replace=False)
    print(f"Tree {tree+1}: Features {sorted(random_features)}")
```
