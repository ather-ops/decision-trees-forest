# XGBoost - Day 6 (Implementation and Hands-on)

## Implementation Overview

Here we implement XGBoost using Python and the Scikit-learn compatible API to train, predict and evaluate a classification model.

---

## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from xgboost import XGBClassifier

%matplotlib inline
sns.set_style("whitegrid")

print("Libraries imported successfully")
print("="*50)
print("XGBoost Day 6 - Implementation")
print("="*50)
```

## Step 2: Load and View the Dataset

```python
df = pd.read_csv("Wholesale customers data.csv")

print("Dataset loaded successfully")
print("="*50)
print("First 5 rows:")
print(df.head())
print("="*50)
print(f"Dataset shape: {df.shape}")
```

## Step 3: Explore Statistical Summary of the Data

```python
print("\nStatistical Summary")
print("="*50)
display(df.describe())
```

## Step 4: Prepare Features and Target, Split Data

```python
X = df.drop('Channel', axis=1)
y = df['Channel'].map({1: 1, 2: 0})

print("Features shape:", X.shape)
print("Target distribution:")
print(y.value_counts())
print("="*50)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"Train size: {len(X_train)}")
print(f"Test size: {len(X_test)}")
```

## Step 5: Build and Train the XGBoost Model

```python
params = {
    'objective': 'binary:logistic',
    'max_depth': 4,
    'learning_rate': 0.1,
    'n_estimators': 100,
    'alpha': 10
}

model = XGBClassifier(**params)

print("Training XGBoost model...")
model.fit(X_train, y_train)
print("Model trained successfully")
print(f"Parameters used: {params}")
```

```python
y_pred = model.predict(X_test)
```

## Step 6: Evaluate Model Accuracy and Performance

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

print("\nClassification Report")
print("="*50)
print(classification_report(y_test, y_pred))
```

## Step 7: Plot Confusion Matrix Heatmap

```python
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## Step 8: Plot Feature Importance

```python
plt.figure(figsize=(10, 8))
xgb.plot_importance(model, importance_type='weight')
plt.title("Feature Importance (Weight)")
plt.tight_layout()
plt.show()
```

```python
# Different importance types
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

xgb.plot_importance(model, importance_type='weight', ax=axes[0])
axes[0].set_title("Importance Type: Weight")

xgb.plot_importance(model, importance_type='gain', ax=axes[1])
axes[1].set_title("Importance Type: Gain")

xgb.plot_importance(model, importance_type='cover', ax=axes[2])
axes[2].set_title("Importance Type: Cover")

plt.tight_layout()
plt.show()
```

## Step 9: Visualize XGBoost Decision Tree

```python
plt.figure(figsize=(20, 12))
xgb.plot_tree(model, num_trees=0, rankdir='TB')
plt.title("XGBoost Decision Tree (Tree 0)")
plt.tight_layout()
plt.show()
```

```python
# Multiple trees visualization
fig, axes = plt.subplots(2, 2, figsize=(20, 16))

for i, ax in enumerate(axes.flat):
    if i < 4:
        xgb.plot_tree(model, num_trees=i, ax=ax, rankdir='TB')
        ax.set_title(f"Tree {i}")

plt.suptitle("XGBoost Decision Trees (First 4 Trees)", fontsize=14)
plt.tight_layout()
plt.show()
```
