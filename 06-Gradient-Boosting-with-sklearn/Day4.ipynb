## What is XGBoost?

XGBoost (eXtreme Gradient Boosting) is an optimized gradient boosting algorithm that combines multiple weak models into a stronger, high-performance model.

**Key Characteristics:**
- Uses decision trees as base learners
- Builds trees sequentially so each tree corrects errors from the previous one (boosting)
- Features parallel processing for faster training on large datasets
- Allows parameter customization to optimize performance for specific problems

---

## How XGBoost Works

It builds decision trees sequentially with each tree attempting to correct the mistakes made by the previous one.

**Step-by-step process:**

1. **Start with a base learner:** The first model (decision tree) is trained on the data. In regression tasks, this base model simply predicts the average of the target variable.

2. **Calculate the errors:** After training the first tree, the errors between predicted and actual values are calculated.

3. **Train the next tree:** The next tree is trained on the errors of the previous tree, attempting to correct them.

4. **Repeat the process:** Continue with each new tree trying to correct errors of previous trees until a stopping criterion is met.

5. **Combine the predictions:** The final prediction is the sum of predictions from all trees.

---

## Mathematics Behind XGBoost Algorithm

It can be viewed as an iterative process where we start with an initial prediction often set to zero. After which each tree is added to reduce errors.

**Model Representation:**

$$\hat{y}_i = \sum_{k=1}^{K} f_k(x_i)$$

Where:
- $\hat{y}_i$ is the final predicted value for the $i^{th}$ data point
- $K$ is the number of trees in the ensemble
- $f_k(x_i)$ represents the prediction of the $k^{th}$ tree for the $i^{th}$ data point

---

## Objective Function

The objective function consists of two parts: a loss function and a regularization term.

$$obj(\theta) = \sum_{i}^{n} l(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)$$

Where:
- $l(y_i, \hat{y}_i)$ is the loss function which computes the difference between true value $y_i$ and predicted value $\hat{y}_i$
- $\Omega(f_k)$ is the regularization term which discourages overly complex trees

---

## Iterative Optimization

Instead of fitting the model all at once, we optimize iteratively. Start with initial prediction $\hat{y}_i^{(0)} = 0$ and at each step add a new tree.

$$\hat{y}_i^{(t)} = \hat{y}_i^{(t-1)} + f_t(x_i)$$

Where:
- $\hat{y}_i^{(t-1)}$ is the prediction from the previous iteration
- $f_t(x_i)$ is the prediction of the $t^{th}$ tree for the $i^{th}$ data point

---

## Regularization Term

The regularization term $\Omega(f_t)$ simplifies complex trees by penalizing the number of leaves and leaf weights:

$$\Omega(f_t) = \gamma T + \frac{1}{2} \lambda \sum_{j=1}^{T} w_j^2$$

Where:
- $T$ is the number of leaves in the tree
- $\gamma$ is a regularization parameter that controls tree complexity
- $\lambda$ is a parameter that penalizes the squared weight of the leaves $w_j$

---

## Information Gain Formula

When deciding how to split nodes, XGBoost computes information gain for every possible split:

$$Gain = \frac{1}{2} \left[ \frac{G_L^2}{H_L + \lambda} + \frac{G_R^2}{H_R + \lambda} - \frac{(G_L + G_R)^2}{H_L + H_R + \lambda} \right] - \gamma$$

Where:
- $G_L, G_R$ are sums of gradients in left and right child nodes
- $H_L, H_R$ are sums of Hessians in left and right child nodes

By calculating information gain for every possible split at each node, XGBoost selects the split that results in the largest gain, effectively reducing errors and improving model performance.

---

## Key Parameters Summary

| Parameter | Description |
|-----------|-------------|
| `n_estimators` | Number of boosting rounds |
| `max_depth` | Maximum tree depth |
| `learning_rate` | Step size shrinkage (eta) |
| `gamma` | Minimum loss reduction for split |
| `reg_alpha` | L1 regularization |
| `reg_lambda` | L2 regularization |
| `subsample` | Fraction of samples per tree |
