# Random Forest - Day 4 (From Scratch)

## Attribute Sampling Definition
Attribute Sampling is a technique where only a random subset of features is tested at each node while building a decision tree.

---

## Core Idea
Instead of checking all features, Random Forest checks only randomly selected features at each split.

---

## Why Attribute Sampling is Used
Attribute sampling increases diversity among decision trees.

---

## Working of Attribute Sampling

Suppose dataset has features: [f1, f2, f3, f4, f5]

At one node, Random Forest may test only: [f1, f3, f5]

Then best split is selected only from these sampled features.

---

## Important Insight
Different nodes test different random features. This creates:
- tree diversity
- independence
- lower overfitting

---

## Attribute Sampling Ratio
Attribute Sampling Ratio determines how many features are tested at each node.

---

## Common Default Values

| Task | Features to use |
|------|-----------------|
| Regression | 1/3 of total features |
| Classification | √(number of features) |

---

## Example

Total features = 9

For classification: √9 = 3

Only 3 random features tested per node.

---

## Main Goal
Reduce correlation between trees.

---

## Why This Helps

If all trees always use same features, then all trees become similar, which reduces Random Forest performance.

---

## Key Insight
- Bagging creates different data for trees
- Attribute sampling creates different feature views for trees

---

## Advantages
- improves diversity
- reduces overfitting
- improves generalization
- creates stronger ensemble

---

## One-Line Summary
**Attribute sampling randomly selects a subset of features at each node for finding the best split.**

---

## Important Terms

- **Feature Sampling:** Random selection of features
- **Split:** Condition used to divide data
- **Correlation:** Similarity between trees

---

**Different trees + different features = stronger forest**

```python
import numpy as np
import math
```

## Random Feature Sampling

```python
features = ["f1", "f2", "f3", "f4", "f5"]

# Randomly select 3 features
sampled_features = np.random.choice(
    features,
    size=3,
    replace=False
)

print("All features:", features)
print("Sampled Features:", sampled_features)
```

## Multiple Nodes Simulation

```python
print("\n" + "="*50)
print("Different Nodes - Different Random Features")
print("="*50)

for node in range(3):
    sampled = np.random.choice(
        features,
        size=3,
        replace=False
    )
    print(f"Node {node+1}:", sampled)
```

## Calculate Features per Split (Classification)

```python
# Example with different total features
total_features_list = [4, 9, 16, 25, 100]

print("\n" + "="*50)
print("Features per Split (Classification)")
print("="*50)

for total in total_features_list:
    features_per_split = int(math.sqrt(total))
    print(f"Total features: {total:3} -> sqrt = {features_per_split:2} features per split")
```
