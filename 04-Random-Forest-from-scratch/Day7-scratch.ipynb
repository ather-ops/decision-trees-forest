## Interpreting Random Forest

Random Forest interpretation means understanding which features and patterns the model uses for prediction.

| Model | Interpretability |
|-------|------------------|
| Single Decision Tree | Easy to visualize |
| Random Forest | Harder due to many trees |

---

## Methods to Interpret Random Forest

### 1. Train a Single CART Tree

Train one decision tree on the same dataset. CART and Random Forest learn similar global patterns.

### 2. Variable Importance

Variable Importance measures how useful a feature is for prediction.

| Method | Meaning |
|--------|---------|
| Sum Score | Total split contribution |
| Num Nodes | How often feature appears |
| Mean Min Depth | How early feature appears |
| Permutation Importance | Performance drop after shuffling |

### 3. Permutation Importance

Measures performance drop after shuffling a feature. If shuffling reduces accuracy heavily, the feature is important.

### 4. SHAP

Explains how much each feature contributes to predictions at the individual and global level.

---

## Random Forest Advantages

- Supports mixed data types (numerical, categorical) without normalization or scaling
- Parallel training (trees are independent)
- Strong default parameters (works well without tuning)

---

## Random Forest Disadvantages

- Large model size (many trees consume memory, slower inference)
- Cannot reuse representations (each tree relearns patterns separately)
- Best on structured tabular data, weaker on images, audio, raw text

---

## Important Terms

- **Model Agnostic:** Works with any ML model
- **Parallel Training:** Multiple trees train simultaneously
- **Internal Representation:** Learned reusable patterns inside a model

---

## One-Line Summary

**Random Forest improves prediction using many independent trees but becomes harder to interpret than a single decision tree.**

## Scratch Idea: Count Feature Usage

```python
import numpy as np
from collections import Counter
```

```python
# Simulated tree structures
trees = [
    {"feature": 0, "left": {"feature": 1}, "right": {"feature": 0}},
    {"feature": 2, "left": {"feature": 0}, "right": {"feature": 3}},
    {"feature": 1, "left": {"feature": 2}, "right": {"feature": 1}},
    {"feature": 0, "left": {"feature": 3}, "right": {"feature": 2}},
]
```

```python
def count_feature_usage(node, counter):
    if isinstance(node, dict):
        if 'feature' in node:
            counter[node['feature']] += 1
            count_feature_usage(node['left'], counter)
            count_feature_usage(node['right'], counter)
    return counter
```

```python
total_count = Counter()
for tree in trees:
    count_feature_usage(tree, total_count)

print("Feature Usage Count Across Trees:")
print("="*40)
for feat, count in sorted(total_count.items()):
    print(f"Feature {feat}: appears {count} times")
```
