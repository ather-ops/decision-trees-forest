# Random Forest - Day 2 (From Scratch)

## Random Forest Definition
Random Forest is an ensemble of decision trees where each tree is trained using random noise and different subsets of data.

---

## Bagging Definition
Bagging (Bootstrap Aggregating) is a technique where each decision tree is trained on a random subset of training examples.

---

## Core Idea of Bagging
Each tree sees different data. This creates:
- diversity
- independence
- reduced overfitting

---

## Bootstrap Sampling
Bootstrap sampling means selecting training examples randomly with replacement.

### With Replacement
With replacement means the same example can be selected multiple times.

**Example:**

Original data: [1,2,3,4,5]

Bootstrap sample: [2,4,4,1,5]
- Example 4 repeated
- Example 3 missing

---

## Important Observation
Each tree is trained on the same total number of samples as the original dataset. BUT:
- Some examples repeat
- Some are absent

---

## Why Bagging Helps

| Single Decision Tree | Bagging |
|----------------------|---------|
| learns exact patterns | different trees learn different patterns |
| overfitting | final averaging reduces variance |

---

## Goal of Bagging
Reduce variance and improve generalization.

---

## With vs Without Replacement

- **With Replacement:** same sample can appear multiple times
- **Without Replacement:** sample appears at most once

---

## Key Insight
Random Forest works well because trees are trained on different data distributions.

---

## One-Line Summary
**Bagging = Random sampling of training data for each decision tree.**

```python
import numpy as np
```

## Bootstrap Sampling Manually

```python
# Original dataset
data = np.array([1,2,3,4,5,6])

# Bootstrap sampling
sample = np.random.choice(data, size=len(data), replace=True)

print("Original:", data)
print("Bootstrap Sample:", sample)
```

## Multiple Trees Simulation

```python
for i in range(3):
    sample = np.random.choice(data, size=len(data), replace=True)
    print(f"Tree {i+1} Training Data:", sample)
```
