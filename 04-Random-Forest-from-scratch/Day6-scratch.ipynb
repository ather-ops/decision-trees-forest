# Random Forest - Day 6 (From Scratch)

## The Clarity of Noise

Random noise improves Random Forest by creating diverse and independent decision trees.

---

## Core Idea

| Single Decision Tree | Random Forest |
|----------------------|---------------|
| learns exact patterns | combines many noisy trees |
| sensitive to noise | averaging removes random errors |
| overfits easily | captures true pattern better |

---

## Important Insight

Individual trees in a Random Forest may perform poorly alone, but together they create strong predictions.

---

## Why Noise Helps

If all trees are identical:
- Ensemble gives no improvement

But if trees are:
- different
- noisy
- weakly correlated

Then their errors cancel each other.

---

## Important Concept

Weakly correlated errors improve ensemble performance.

---

## Why Unpruned Trees Work

Random Forest uses:
- bagging
- attribute sampling
- deep trees

These create:
- randomness
- diversity
- independence

Therefore: Overfitting of one tree is corrected by other trees.

---

## Adding More Trees

Important property: Adding more trees usually improves or stabilizes performance.

---

## Random Forest Overfitting

Increasing the number of trees generally does not cause overfitting.

Accuracy:
- improves rapidly initially
- eventually plateaus
- rarely decreases

---

## Leo Breiman Statement

"Random Forests do not overfit as more trees are added."

---

## Why Accuracy Stabilizes

As trees increase:
- variance decreases
- predictions stabilize
- ensemble becomes robust

---

## One-Line Summary

**Random Forest improves performance by combining many noisy and independent decision trees.**

---

## Advantages

- lower variance
- reduced overfitting
- robust predictions
- strong generalization

---

## Important Terms

- **Noise:** Random variations in predictions
- **Correlated Errors:** Trees making similar mistakes
- **Weakly Correlated Errors:** Trees making different mistakes
- **Ensemble Averaging:** Combining multiple predictions

---

**Many noisy independent trees combine to create a strong Random Forest.**

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Simulating Ensemble Averaging

```python
# True value
true_value = 50

# Noisy predictions from trees
tree_predictions = np.array([45, 55, 48, 52, 49])

# Ensemble prediction
final_prediction = np.mean(tree_predictions)

print("True Value:", true_value)
print("Tree Predictions:", tree_predictions)
print("Ensemble Prediction:", final_prediction)
print(f"Error reduced from {np.abs(tree_predictions - true_value).mean():.1f} to {np.abs(final_prediction - true_value):.1f}")
```

## Visual Noise Effect

```python
# True curve
x = np.linspace(0, 10, 100)
y_true = np.sin(x)

plt.figure(figsize=(12, 6))

# Plot noisy trees
for i in range(5):
    noise = np.random.normal(0, 0.3, size=len(x))
    y_noisy = y_true + noise
    plt.plot(x, y_noisy, alpha=0.5, label=f'Tree {i+1}' if i == 0 else "")

# Plot true pattern
plt.plot(x, y_true, linewidth=3, color='black', label='True Pattern')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Noisy Trees vs True Pattern")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Show how ensemble average approximates true pattern
np.random.seed(42)

n_trees = 20
all_predictions = []

for i in range(n_trees):
    noise = np.random.normal(0, 0.3, size=len(x))
    y_noisy = y_true + noise
    all_predictions.append(y_noisy)

all_predictions = np.array(all_predictions)
ensemble_avg = np.mean(all_predictions, axis=0)

plt.figure(figsize=(12, 6))
plt.plot(x, y_true, 'k-', linewidth=3, label='True Pattern')
plt.plot(x, ensemble_avg, 'r--', linewidth=2, label=f'Ensemble Average ({n_trees} trees)')
plt.fill_between(x, y_true - 0.3, y_true + 0.3, alpha=0.2, label='Noise Band')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ensemble Averaging Removes Noise")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
