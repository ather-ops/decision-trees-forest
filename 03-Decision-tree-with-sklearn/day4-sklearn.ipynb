## Decision Tree Training

Decision tree is built using a recursive function that:
- Finds best condition
- Splits data
- Repeats until no split is possible

The algorithm then repeats recursively and independently on both children nodes. When no satisfying conditions are found, the node becomes a leaf. The leaf prediction is determined as the most representative label value in the examples.

---

## Training Algorithm Pseudocode

```
def train_decision_tree(training_examples):
    root = create_root()
    grow_tree(root, training_examples)
    return root

def grow_tree(node, examples):
    condition = find_best_condition(examples)

    if condition is None:
        set_leaf_prediction(node, examples)
        return

    positive_child, negative_child = split_node(node, condition)

    negative_examples = [example for example in examples if not condition(example)]
    positive_examples = [example for example in examples if condition(example)]

    grow_tree(negative_child, negative_examples)
    grow_tree(positive_child, positive_examples)
```

---

## Step-by-Step Training Process

**Step 1:** Create a root node with a question mark

**Step 2:** Grow node #1. Find condition (example: x1 >= 1). Create two child nodes

**Step 3:** Grow node #2. If no satisfying conditions found, make the node a leaf

**Step 4:** Grow node #3. Find condition (example: x2 >= 0.5). Create two child nodes

---

## Important Concepts

**Splitter:** The routine responsible for finding the best condition. Splitters are the bottleneck when training a decision tree because they need to test many possible conditions.

**Score metrics depend on task:**
- Information gain and Gini for classification
- Mean squared error for regression

**Splitter considerations:**
- Type of features (numerical, categorical, text)
- Type of task (binary classification, multi-class, regression)
- Type of condition (threshold, in-set, oblique)
- Regularization criteria
- Memory and CPU trade-offs

````python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
````

## Training From Scratch vs Scikit-Learn

````python
# Dataset
X = np.array([0, 1, 2, 3, 4]).reshape(-1, 1)
y = np.array(["A", "A", "B", "B", "B"])

# Scikit-Learn Decision Tree
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)

print("Training completed!")
print(f"Tree depth: {clf.tree_.max_depth}")
print(f"Number of nodes: {clf.tree_.node_count}")
print(f"Number of leaves: {clf.tree_.n_leaves}")
````

````python
# Visualize the trained tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['Feature'], 
          class_names=['A', 'B'], 
          filled=True, rounded=True)
plt.title("Trained Decision Tree")
plt.show()
````

````python
# Make predictions
X_test = np.array([0.5, 1.5, 2.5, 3.5, 5]).reshape(-1, 1)
predictions = clf.predict(X_test)

print("Predictions on new values:")
print("="*30)
for val, pred in zip(X_test, predictions):
    print(f"{val[0]} -> {pred}")
````

## Understanding the Splitter in Scikit-Learn

````python
# Different splitter options
print("Splitter Options in Scikit-Learn:")
print("="*50)

# 'best' splitter - tries all features
clf_best = DecisionTreeClassifier(splitter='best', max_depth=2)
clf_best.fit(X, y)
print("1. splitter='best': Tries all features and thresholds")

# 'random' splitter - random selection
clf_random = DecisionTreeClassifier(splitter='random', max_depth=2)
clf_random.fit(X, y)
print("2. splitter='random': Selects random split")

print("="*50)
````

````python
# Compare different criteria
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gini criterion
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=2)
clf_gini.fit(X, y)
plot_tree(clf_gini, feature_names=['Feature'], class_names=['A', 'B'], 
          filled=True, rounded=True, ax=axes[0])
axes[0].set_title("Gini Impurity")

# Entropy criterion
clf_entropy = DecisionTreeClassifier(criterion='entropy', max_depth=2)
clf_entropy.fit(X, y)
plot_tree(clf_entropy, feature_names=['Feature'], class_names=['A', 'B'], 
          filled=True, rounded=True, ax=axes[1])
axes[1].set_title("Entropy")

plt.suptitle("Comparing Split Criteria", fontsize=14)
plt.tight_layout()
plt.show()
````
