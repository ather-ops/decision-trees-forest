## Growing Decision Tree

Growing a decision tree is the process of building the tree by recursively splitting the dataset into smaller subsets using the best possible conditions.

A decision tree learns patterns by dividing data into smaller and more homogeneous (pure) groups.

**Pure means:** Same class in one group

---

## Solution: Heuristic Approach

Decision trees are trained using heuristics, which are approximate methods that produce near-optimal solutions efficiently.

**Example:**
- Instead of checking 100,000 options
- Pick the best option at each step

---

## Algorithm Used

Decision trees use a greedy divide-and-conquer strategy.

### Divide and Conquer
- Divide data into smaller parts
- Solve each part separately

### Greedy Approach
- At each step, the algorithm selects the best split based only on current data
- Does not think ahead
- No future planning

---

## Step-by-Step Process

1. Start with entire dataset (root node)
2. Try all possible conditions
3. Evaluate each condition using a score
4. Select the best condition
5. Split dataset into two parts
6. Repeat the process for each part
7. Stop when no good split is possible

---

## What is "Best Condition"?

Best condition is the split that gives the highest score (i.e., best separation of data).

**Score measures:** How well classes are separated

(Gini / Entropy)

---

## Example (Clear Understanding)

**Dataset:**

| Species   | Bill Length |
|-----------|-------------|
| Adelie    | 18          |
| Chinstrap | 17          |
| Gentoo    | 14          |
| Gentoo    | 13          |

**Condition:** bill_length >= 16

**Result:**

| Condition | Output                    |
|-----------|---------------------------|
| Yes (>=16)| Adelie, Chinstrap         |
| No (<16)  | Gentoo, Gentoo            |

---

## Recursive Growth

After the first split, the algorithm repeats the process independently on each child node.

**Example:**

**Step 1:**
Split -> [Adelie, Chinstrap] | [Gentoo]

**Step 2:**
Now split only left part -> [Adelie] | [Chinstrap]

---

## When Does Tree Stop?

A node becomes a leaf when no further useful split is possible.

**Conditions:**
- Data is pure
- No good split found
- Stopping rule reached

---

## Leaf Node Prediction

Leaf node stores the final prediction based on the data it contains.

**Rule:** Prediction = most frequent class (majority vote)

**Example:** ["dog", "dog", "cat"] -> dog

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

## Example Dataset - Penguin Species

```python
# Dataset: Bill Length feature only
X = np.array([[18], [17], [14], [13]])
y = np.array(['Adelie', 'Chinstrap', 'Gentoo', 'Gentoo'])

# Train Decision Tree
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)

# Visualize the tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['Bill Length'], 
          class_names=['Adelie', 'Chinstrap', 'Gentoo'], 
          filled=True, rounded=True)
plt.title("Decision Tree - Penguin Species Classification")
plt.show()
```

```python
# Print tree structure
print("Tree Structure:")
print("="*40)
print(f"Root node splits on Bill Length <= {clf.tree_.threshold[0]:.2f}")
print(f"  Left child (<= threshold): {clf.tree_.children_left[0]}")
print(f"  Right child (> threshold): {clf.tree_.children_right[0]}")
print("="*40)
```

## Step-by-Step Growing Process Visualization

```python
# Show how tree grows with different depths
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for depth, ax in zip([1, 2, 3], axes):
    clf_depth = DecisionTreeClassifier(max_depth=depth)
    clf_depth.fit(X, y)
    plot_tree(clf_depth, feature_names=['Bill Length'], 
              class_names=['Adelie', 'Chinstrap', 'Gentoo'],
              filled=True, rounded=True, ax=ax)
    ax.set_title(f"Depth = {depth}")

plt.suptitle("Tree Growth with Increasing Depth", fontsize=14)
plt.tight_layout()
plt.show()
```

## Stopping Conditions in Scikit-Learn

```python
# Different stopping conditions
print("Stopping Conditions Examples:")
print("="*50)

# 1. Max Depth
clf_depth = DecisionTreeClassifier(max_depth=1)
clf_depth.fit(X, y)
print(f"1. Max Depth=1: Tree stopped at depth 1")

# 2. Min Samples Split
clf_min_samples = DecisionTreeClassifier(min_samples_split=3)
clf_min_samples.fit(X, y)
print(f"2. Min Samples Split=3: Won't split nodes with <3 samples")

# 3. Min Samples Leaf
clf_min_leaf = DecisionTreeClassifier(min_samples_leaf=2)
clf_min_leaf.fit(X, y)
print(f"3. Min Samples Leaf=2: Leaves must have at least 2 samples")

print("="*50)
```

## Leaf Node Predictions

```python
# Show leaf node predictions
X_test = np.array([[15], [16], [17], [18]])
predictions = clf.predict(X_test)
probabilities = clf.predict_proba(X_test)

print("Leaf Node Predictions:")
print("="*50)
for i, (val, pred, prob) in enumerate(zip(X_test, predictions, probabilities)):
    print(f"Bill Length {val[0]}: Predicted = {pred}")
    print(f"  Probabilities: Adelie={prob[0]:.2f}, Chinstrap={prob[1]:.2f}, Gentoo={prob[2]:.2f}")
    print("-"*40)
```
