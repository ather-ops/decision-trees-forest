## Decision Forest
A decision forest (Random Forest) is an ensemble learning method used for classification and regression that operates by constructing multiple Decision Trees during training time.

## Decision Tree
A decision tree is a supervised machine learning model used for classification and regression that splits data into smaller subsets based on feature conditions.

A decision tree is a model composed of a collection of "questions" organized hierarchically in the shape of a tree. The questions are usually called **Conditions**.

> Each non-leaf node contains a condition, and each leaf contains a prediction.

Botanical trees generally grow with the root; however, decision trees are usually represented with the root (the first node) at the top. It is structured like a tree.

## Key Terminology

**Root Node:** First decision. Topmost node. Represents entire dataset. First split happens here.

**Decision Node (Internal Nodes):** Conditions (questions). A node where splitting occurs. Contains a condition.

**Leaf Node (Terminal Node):** Final node. Contains prediction (class or value).

**Branch:** Outcome of a decision. Connects nodes.

**Splitting:** Process of dividing data based on feature condition.

**Depth:** Number of levels in tree.

## Working Principle

1. Start with full dataset
2. Try all features and split points
3. Select best split using impurity measure
4. Divide dataset
5. Repeat recursively
6. Stop when stopping condition met

## Impurity Measures

**Gini Impurity (used in CART)**
Gini = 1 - sum(p_i^2)
Measures how impure a node is. Value range: 0 (pure) -> 0.5 (impure for binary)

**Entropy (Information Theory)**
Entropy = -sum(p_i * log2(p_i))
Measures randomness. Higher entropy = more disorder.

**Information Gain**
Information Gain = Parent Entropy - Weighted Child Entropy
Goal: maximize information gain

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

## Tree Example

IF num_legs >= 3:
    IF num_eyes >= 3:
        -> spider
    ELSE:
        -> dog
ELSE:
    -> penguin

```python
# Example input: [num_legs, num_eyes]
X = np.array([
    [4, 2],
    [8, 8],
    [2, 2]
])

y = np.array(['dog', 'spider', 'penguin'])

# Train Decision Tree
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)

# Predict
for x in X:
    pred = clf.predict([x])[0]
    print(x, "->", pred)
```

```python
# Visualize the tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['num_legs', 'num_eyes'], 
          class_names=['dog', 'penguin', 'spider'], filled=True)
plt.title("Decision Tree Structure")
plt.show()
```

```python
# Test on new samples
new_samples = np.array([[6, 2], [10, 10], [2, 4]])
print("\nPredictions on new samples:")
for x in new_samples:
    pred = clf.predict([x])[0]
    print(x, "->", pred)
```
